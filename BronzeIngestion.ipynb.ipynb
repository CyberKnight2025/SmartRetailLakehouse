```python
# Welcome to your new notebook
# Bronze Layer Ingestion for Sales Data

# Step 1: Read CSV from Lakehouse 'Files/raw/'
df_sales = spark.read.option("header", "true").csv("Files/raw/sales.csv")

# Step 2: Display schema and preview rows
df_sales.printSchema()
df_sales.show(5)

# Step 3: Clean minimal nulls
df_sales_clean = df_sales.dropna(subset=["Item", "Quantity", "UnitPrice"])

# Step 4: Write to Bronze Delta table
df_sales_clean.write.format("delta").mode("overwrite").save("Tables/Bronze_Sales")

print("Bronze Sales Table created successfully.")
```
