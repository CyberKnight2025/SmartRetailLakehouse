```python
# Welcome to your new notebook
# Silver Layer Transformation - Sales Data

# Step 1: Read from Bronze Delta Table
df_bronze = spark.read.format("delta").load("Tables/Bronze_Sales")

# Step 2: Display schema and preview rows
df_bronze.printSchema()
df_bronze.show(5)

# Step 3: Convert datatypes
df_silver = df_bronze.withColumn("Quantity", df_bronze["Quantity"].cast("int")) \
                     .withColumn("UnitPrice", df_bronze["UnitPrice"].cast("double")) \
                     .withColumn("TaxAmount", df_bronze["TaxAmount"].cast("double"))

# Step 4: Add TotalPrice column (Quantity * UnitPrice)
df_silver = df_silver.withColumn("TotalPrice", df_silver["Quantity"] * df_silver["UnitPrice"])

# Step 5: Preview the result
df_silver.printSchema()
df_silver.show(5)

# Step 6: Write to Silver table
df_silver.write.format("delta").mode("overwrite").save("Tables/Silver_Sales")

print("Silver Sales Table created successfully.")
```
